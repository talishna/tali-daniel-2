## p-wave superconductor (Kitaev toy model)

$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu  \hat{c}_{j}^\dagger \hat{c}_{j} -\left( \frac{t}{2} \left( \hat{c}^\dagger_{j} \hat{c}_{j+1} + \hat{c}_{j+1}^\dagger\hat{c}_{j} \right)  +\frac{\Delta}{2} \left( \hat{c}^\dagger_{j} \hat{c}^\dagger_{j+1} + \hat{c}_{j+1} \hat{c}_{j} \right) \right) \right]
$$
We will use open boundary conditions and switch to the Majorana representation.

### Majorana operator
$$
\hat{c}= \frac{1}{2} \left(\gamma_{b,j} + i\gamma_{a,j} \right)
$$
Where $ \gamma_{\alpha}$ are the Majorana operators.

#### Majorana algebra:
The Majorana operators have the following algebra:

\begin{align}
\gamma_{\alpha}^\dagger &= \gamma_{\alpha} \\
\gamma_{\alpha}^2&= \hat{\mathcal{I}}\\
\left\{ \gamma_{\alpha},\gamma_{\alpha'\neq \alpha} \right\} &= 0
\end{align}

The above algebra reproduces the usual fermion algebra:

$$
\hat{c}^2 = \left( \hat{c}^\dagger \right) ^2 \\
\left\{ \hat{c}_{j}, \hat{c}_{j'\neq j} \right\} = \delta_{jj'} \\
$$

We will look at the case where $\Delta = t $:

$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu  \hat{c}_{j}^\dagger \hat{c}_{j} - \frac{t}{2}\left(\hat{c}_{j}^\dagger + \hat{c}_{j}\right) \left( \hat{c}_{j+1} - \hat{c}_{j+1}^\dagger \right) \right]
$$

Now let's rewrite the above Hamiltonian (represent the Hamiltonian in the Majorana operators):

$$
\hat{c}_{j}^\dagger \hat{c}_{j} = \frac{1}{4} \left(\gamma_{b,j} + i\gamma_{a,j} \right) \left(\gamma_{b,j} - i\gamma_{a,j} \right) = \frac{1}{4} \left( \gamma_{b,j}^2 + i\gamma_{b,j}\gamma_{a,j} - i\gamma_{a,j}\gamma_{b,j} + \gamma_{a,j}^2 \right) = \frac{1}{4} \left( 1 + i\gamma_{b,j}\gamma_{a,j} + i\gamma_{b,j}\gamma_{a,j} + 1 \right) =  \frac{1}{2} \left(1 + i\gamma_{b,j}\gamma_{a,j} \right) \\
\left(\hat{c}_{j}^\dagger + \hat{c}_{j}\right) \left(\hat{c}_{j+1} - \hat{c}_{j+1}^\dagger\right) = i\gamma_{b,j}\gamma_{a,j+1} \\
\hat{\mathcal{H}} = \frac{-i}{2} \sum_{j} \left[ \mu \gamma_{b,j} \gamma_{a,j} + t \gamma_{b,j} \gamma_{a,j+1} \right] \\
$$


# Jordan-Wigner transformation 
The Hamiltonian now is:
$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu  \hat{c}_{j}^\dagger \hat{c}_{j} - \frac{t}{2} \left( \hat{c}^\dagger_{j} \hat{c}_{j+1} + \hat{c}_{j+1}^\dagger\hat{c}_{j}  \right) \right]
$$
The Jordan-Wigner Transformation is a transformation from the Spin basis to the Fermion basis, we'll want to do the opposite transformation:
$$
\hat{c}_{j}^{\dagger} = \hat{K}(j) \hat{S}_{j}^{+}  \\
\hat{c}_{j} = \hat{S}_{j}^{-} \hat{K}(j)   \\
\hat{c}_{j}^{\dagger} \hat{c}_{j} = \hat{S}_{j}^{z} + \frac{1}{2} \\
$$
Where $\hat{K}(j)$ is defined as:
$$
\hat{K}(j) = \exp\biggl(\, i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \exp\left(\, i\pi \sum_{k=1}^{j-1}  \left( \hat{S}_{j}^{z} + \frac{1}{2} \right) \right) = \prod_{k=1}^{j-1} \left( -\sigma_{j}^{z} \right) \\
\hat{K}(j+1) = \hat{K}(j)(-\sigma_{j}^{z})
$$
Because $\hat{c}_{j}^{\dagger}\hat{c}_{j}$ is the fermionic occupation number it has value of 0 or 1, so:
$$
\hat{K}(j) = \exp\biggl(\, i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \exp\biggl(\, -i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \hat{K}(j)^{\dagger} \\
\left[ \hat{K}(i), \hat{K}(j) \right] = 0
$$
Furthermore for $i \leq j$ :
$$
\left[ \hat{K}(i), \hat{S}_{j} \right] = 0
$$
We should notice that:
$$
\hat{c}^\dagger_{j} \hat{c}_{j+1} = \hat{K}(j) \hat{S}_{j}^{+} \hat{S}_{j+1}^{-} \hat{K}(j+1) = \hat{S}_{j}^{+}\hat{K}(j) \hat{K}(j+1) \hat{S}_{j+1}^{-} = \hat{S}_{j}^{+} (-\sigma_{j}^{z}) \hat{S}_{j+1}^{-} = \hat{S}_{j}^{+} \hat{S}_{j+1}^{-}
$$
Where the last equality is correct because if the spin in the jth site is up then $\hat{S}_{j}^{+}$ will destroy the state. And if the spin is down then $-\sigma_{j}^{z}=1$.
Let's rewrite the Hamiltonian:
$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu \left( \hat{S}_{j}^{z} + \frac{1}{2} \right) - t \left( \hat{S}_{j}^{+} \hat{S}_{j+1}^{-} + \hat{S}_{j+1}^{+} \hat{S}_{j}^{-}  \right) \right]
$$
Where $ \hat{S}_{j}^{+} , \hat{S}_{j}^{-}$ are the usual creation and annihilation (raising and lowering) operators:
$$
\hat{S}_{j}^{+} = \hat{S}_{j}^{x} + i\hat{S}_{j}^{y} \\
\hat{S}_{j}^{-} = \hat{S}_{j}^{x} - i\hat{S}_{j}^{y}
$$
And $ \hat{S}^{x}, \hat{S}^{y} , \hat{S}^{z} $ are the usual Pauli matrices:

$$\hat{S}^{x} = \frac{1}{2} \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$
$$\hat{S}^{y} = \frac{1}{2} \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}$$
$$\hat{S}^{z} = \frac{1}{2} \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

# Defining the single spin operators s_x, s_y, s_z:
s_x = 0.5*np.array([[0,1],[1,0]], dtype=complex)
s_y = 0.5*np.array([[0,-1j],[1j,0]], dtype=complex)
s_z = 0.5*np.array([[1,0],[0,-1]], dtype=complex)
N = 3 #number of spins
s_plus = s_x + 1j*s_y
s_minus = s_x - 1j*s_y

# Defining the N spins operators
def s(matrix, spin_index): # matrix is the matrix I want to do kron prod for. options are: s_z, s_plus, s_minus # spin_index is the spin index inside the chain, so if it's 1 it'll be the 1st spin if 2 the 2nd spin ect.
    if spin_index == 1: # dealing with the case that there is no need for identity matrix on the left side which is only for the 1st spin
        M = np.kron(matrix, np.identity(2**(N-1), dtype=complex))
    elif spin_index == N: # dealing with the case that there is no need for identity matrix on the right side which is only for the Nth spin
        M = np.kron(np.identity(2**(spin_index-1), dtype=complex),matrix)
    else: # for all other cases where 1<index<N and then there is need for identity matrices on both sides
        M = np.kron(np.identity(2**(spin_index-1), dtype=complex),np.kron(matrix, np.identity(2**(N-spin_index), dtype=complex))) # the first identity matrix is of size 2^i the last identity matrix is of size N-i
    return M

def k(spin_index): # calculates k(i) for the jordan-wigner trans
    matrix = np.identity(2**N, dtype=complex)
    for i in range(1, spin_index): # this reaches i-1, and I used the last equality for k(i)
        matrix = np.matmul(matrix, -2*s(s_z, i))
    return matrix

def H(mu,t): # calculates the hamiltonian
    H = np.zeros((2**N,2**N), dtype=complex)
    for i in range(1,N): # this is for the N-1 first objects in the sum of H
        H += -mu*(s(s_z,i)+0.5)-t*(np.matmul(k(i),np.matmul(s(s_plus,i),np.matmul(s(s_minus,i+1),k(i+1)))) + np.matmul(k(i+1),np.matmul(s(s_plus,i+1),np.matmul(s(s_minus,i),k(i)))))
    H += -mu*(s(s_z,N)+0.5) # this is the last object for which the t term is zero because i+1>N (the boundary is open)
    return H

## test if the S function works proparly
# print(s(s_x,2))
# I = np.array([[1,0],[0,1]])
# test = np.kron(I,np.kron(s_x,I))
# np.array_equal(s(s_x,2), test)

## test H
# t=1
# print(H(0.1,t))

# now I need to calc H for different mu and find the eigenvalues(energies) and plot energy for different mu
number_of_mus = 10
mu = np.arange(0, number_of_mus, 1)
H_tot = np.zeros(number_of_mus, dtype=complex)
a = 5
```
