## t' model 
We consider the following model
$$
\hat{\mathcal{H}}=\hat{\mathcal{H}}_L+\hat{\mathcal{H}}_R+\hat{\mathcal{H}}_J
$$
where $L$ and $R$ is our left and right reservoirs, $L\in \left\{1, \ldots, N\right\}$ , $R\in\left\{N+2, \ldots, 2N+1\right\}$: 
$$ 
\hat{\mathcal{H}}_{\alpha}=-\mu_\alpha \sum_{j}\hat{c}_{j}^\dagger \hat{c}_{j} - t\sum_{j} \left( \hat{c}^\dagger_{j} \hat{c}_{j+1} + \hat{c}_{j+1}^\dagger\hat{c}_{j} \right)
$$
with $\alpha\in\{L,R\}$, and $J$ is our junction: 
$$
\hat{\mathcal{H}}_{J}=-\mu_{J}\hat{c}_{N+1}^\dagger \hat{c}_{N+1} - t'\left( \hat{c}^\dagger_{N} \hat{c}_{N+1} + \hat{c}_{N+1}^\dagger\hat{c}_{N+2}\right)+\mathrm{h.c}
$$
### Task: do a Jordan Wigner Transformation to the Hamiltonian and simplify it.
## Current operator:
The current operator is defined by the rate of change in particles from each side:
$$
\hat{I}_{j,j+1}= e \frac{d}{dt}[N_{L}-N_{R}]=\ldots=i\left(t_{j,j+1}\hat{c}^\dagger_{j} \hat{c}_{j+1}-t_{j,j+1}^*\hat{c}_{j+1}^\dagger\hat{c}_{j}   \right)
$$

### Task: fill in the derivation for a general nearest neighbor particle conserving model, i.e, the model contains hopping between next sites with arbitrary,  nearest neighbor interaction terms, and chemical potencial
### Task: preform a Jordan Wigner transformation to the current operator and simplify it.

## Kondo model + spinless interaction

$$
H_0=H_{spinfull} + H_{dot}+H_{spinless}+H_{tun}+H_{int}
$$
$$

$$
H_{spinless}=-\mu \sum_{j}\hat{f}_{j}^\dagger \hat{f}_{j} - t\sum_{j}  \hat{f}^\dagger_{j} \hat{f}_{j+1} + \hat{f}_{j+1}^\dagger\hat{f}_{j}

$$

$$
H_{V}=\frac{V}{2} \left(\sum_{j\in L}\hat{f}_{j}^\dagger \hat{f}_{j}-\sum_{j\in R}\hat{f}_{j}^\dagger \hat{f}_{j} \right)

where 
$$
H(t)=\begin{cases}

$$


# Jordan-Wigner transformation 
The Hamiltonian now is:
$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu  \hat{c}_{j}^\dagger \hat{c}_{j} - \frac{t}{2} \left( \hat{c}^\dagger_{j} \hat{c}_{j+1} + \hat{c}_{j+1}^\dagger\hat{c}_{j}  \right) \right]
$$
The Jordan-Wigner Transformation is a transformation from the Spin basis to the Fermion basis, we'll want to do the opposite transformation:
$$
\hat{c}_{j}^{\dagger} = \hat{K}(j) \hat{S}_{j}^{+}  \\
\hat{c}_{j} = \hat{S}_{j}^{-} \hat{K}(j)   \\
\hat{c}_{j}^{\dagger} \hat{c}_{j} = \hat{S}_{j}^{z} + \frac{1}{2} \\
$$
Where $\hat{K}(j)$ is defined as:
$$
\hat{K}(j) = \exp\biggl(\, i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \exp\left(\, i\pi \sum_{k=1}^{j-1}  \left( \hat{S}_{j}^{z} + \frac{1}{2} \right) \right) = \prod_{k=1}^{j-1} \left( -\sigma_{j}^{z} \right) \\
\hat{K}(j+1) = \hat{K}(j)(-\sigma_{j}^{z})
$$
Because $\hat{c}_{j}^{\dagger}\hat{c}_{j}$ is the fermionic occupation number it has value of 0 or 1, so:
$$
\hat{K}(j) = \exp\biggl(\, i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \exp\biggl(\, -i\pi \sum_{k=1}^{j-1} \hat{c}_{J}^{\dagger} \hat{c}_{j} \biggr) = \hat{K}(j)^{\dagger} \\
\left[ \hat{K}(i), \hat{K}(j) \right] = 0
$$
Furthermore for $i \leq j$ :
$$
\left[ \hat{K}(i), \hat{S}_{j} \right] = 0
$$
We should notice that:
$$
\hat{c}^\dagger_{j} \hat{c}_{j+1} = \hat{K}(j) \hat{S}_{j}^{+} \hat{S}_{j+1}^{-} \hat{K}(j+1) = \hat{S}_{j}^{+}\hat{K}(j) \hat{K}(j+1) \hat{S}_{j+1}^{-} = \hat{S}_{j}^{+} (-\sigma_{j}^{z}) \hat{S}_{j+1}^{-} = \hat{S}_{j}^{+} \hat{S}_{j+1}^{-}
$$
Where the last equality is correct because if the spin in the jth site is up then $\hat{S}_{j}^{+}$ will destroy the state. And if the spin is down then $-\sigma_{j}^{z}=1$.
Let's rewrite the Hamiltonian:
$$
\hat{\mathcal{H}}= \sum_{j} \left[ -\mu \left( \hat{S}_{j}^{z} + \frac{1}{2} \right) - t \left( \hat{S}_{j}^{+} \hat{S}_{j+1}^{-} + \hat{S}_{j+1}^{+} \hat{S}_{j}^{-}  \right) \right]
$$
Where $ \hat{S}_{j}^{+} , \hat{S}_{j}^{-}$ are the usual creation and annihilation (raising and lowering) operators:
$$
\hat{S}_{j}^{+} = \hat{S}_{j}^{x} + i\hat{S}_{j}^{y} \\
\hat{S}_{j}^{-} = \hat{S}_{j}^{x} - i\hat{S}_{j}^{y}
$$
And $ \hat{S}^{x}, \hat{S}^{y} , \hat{S}^{z} $ are the usual Pauli matrices:

$$\hat{S}^{x} = \frac{1}{2} \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$
$$\hat{S}^{y} = \frac{1}{2} \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}$$
$$\hat{S}^{z} = \frac{1}{2} \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as linalg
from scipy import sparse as sparse
import scipy.sparse.linalg as sp_linalg
import math


# Defining the single spin operators s_x, s_y, s_z:
s_x = 0.5*sparse.csc_matrix(np.array([[0,1],[1,0]], dtype=complex))
s_y = 0.5*sparse.csc_matrix(np.array([[0,-1j],[1j,0]], dtype=complex))
s_z = 0.5*sparse.csc_matrix(np.array([[1,0],[0,-1]], dtype=complex))
N = 10 #number of spins - keep it an even number for simplicity
fill = np.floor(N/2) #the
s_plus = s_x + 1j*s_y
s_minus = s_x - 1j*s_y

# Defining the N spins operators
def s(matrix, spin_index): # matrix is the matrix I want to do kron prod for. options are: s_z, s_plus, s_minus # spin_index is the spin index inside the chain, so if it's 1 it'll be the 1st spin if 2 the 2nd spin ect.
    M = sparse.kron(sparse.identity(2**(spin_index-1), dtype=complex),sparse.kron(matrix, sparse.identity(2**(N-spin_index), dtype=complex))) # the first identity matrix is of size 2^i the last identity matrix is of size N-i
    return M

def k(spin_index): # calculates k(i) for the jordan-wigner trans
    matrix = sparse.identity(2**N, dtype=complex)
    for i in range(1, spin_index): # this reaches i-1, and I used the last equality for k(i)
        matrix = matrix.dot(-2*s(s_z, i))
    return matrix

def H0(mu,t): # calculates the hamiltonian
    H = sparse.csc_matrix((2**N,2**N), dtype=complex)
    for i in range(1,N): # this is for the N-1 first objects in the sum of H
        H += -mu*(s(s_z+0.5*sparse.identity(2, dtype=complex),i))\
             -t*(s(s_plus,i).dot(s(s_minus,i+1)) + s(s_plus,i+1).dot(s(s_minus,i)))
    H += -mu*(s(s_z+0.5*sparse.identity(2, dtype=complex),N)) # this is the last object for which the t term is zero because i+1>N and the boundary is open
    return H

def H1(mu_right,t): # calculates the hamiltonian of the t' model, mu_right is the potential on the right side of the chain and mu_left is 0
    H = sparse.csc_matrix((2**N,2**N), dtype=complex)
    for i in range(1,int(N/2+1)): # this is for the N-1 first objects in the sum of H
        H += -t*(s(s_plus,i).dot(s(s_minus,i+1)) + s(s_plus,i+1).dot(s(s_minus,i)))
    for i in range(int(N/2+1),N): # this is for the N-1 first objects in the sum of H
        H += -mu_right*(s(s_z+0.5*sparse.identity(2, dtype=complex),i))\
             -t*(s(s_plus,i).dot(s(s_minus,i+1)) + s(s_plus,i+1).dot(s(s_minus,i)))
    H += -mu_right*(s(s_z+0.5*sparse.identity(2, dtype=complex),N)) # this is the last object for which the t term is zero because i+1>N and the boundary is open
    return H

def Sz_Total(): # calculates the total number of spins operator
    SzT = sparse.csc_matrix((2**N,2**N), dtype=complex)
    for i in range(1,N+1): # this is for the N-1 first objects in the sum of H
        SzT += s(s_z+0.5*sparse.identity(2, dtype=complex),i)
    #SzT += s(s_z+0.5*sparse.identity(2, dtype=complex),N)# this is the last object for which the t term is zero because i+1>N and the boundary is open
    return SzT

constant_of_motion=np.diag(Sz_Total().toarray()) # can also use sparse.diags
```

```python
## our H in t<0 is H0 we then turn on voltage by applying H1 for t>0 for which we find the eigenvectors and eigenvalues
## we then calculate the current expectation value
t=0.5
dt = 0.01
mu0 = 0
mu1 = 0.1
Ham0=H0(mu0,t) #GS in t<0
Hamt=H1(mu1,t) #H in time t
#Hamt=sparse.csc_matrix(Hamt)
Udt = sparse.linalg.expm(-1j*Hamt*dt) # Evolution operator
[energy,y] = linalg.eigh(Ham0.toarray()) # Out eigen stuff for H0 (for t<0)
print("found energy and vectors")
vecGS = y[:,0] # This is the ground state vector which I won't touch and will only be used for the fidelity test
vec = y[:,0] # The vector that will be forwarded in time
current = 1j*((t*s(s_plus,5).dot(s(s_minus,6)))-(t*s(s_minus,5).dot(s(s_plus,6)))) # Current operator between sites 5 and 6
ExpValCurrent = np.zeros(int((1/dt)*N),dtype=complex) # The expectation value of the current , here I used lil_matrix because I will add/remove entries and csc_matrix is not a good tool in this case - adding/removing entries in it is expensive computationally
fidelity = ExpValCurrent
for i in range(0,int((1/dt)*N)):
    ExpValCurrent[i] = vec.conjugate().dot(current.dot(vec))
    fidelity[i] = np.abs(vecGS.conjugate().dot(vec))
    vec = Udt.dot(vec)
print(vecGS)
```
