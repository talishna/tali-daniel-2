```python

def H_Transmon(local_E_C = E_C, local_n_0 = n_0, local_n_g = n_g, local_E_J_max = E_J_max, local_d = d, flux=0, local_cutoff_transmon = False, local_size_of_subspace_T = None): #N is the number of cp passed from one sc to the other
    # when cutoff = false I return H,eigenvalues,eigenvectors in the size of the whole transmon space
    # when cutoff = true I return H,eigenvalues,eigenvectors. H in the size of the subspace and eigenvalues,eigenvectors in the size of the whole transmon space
    D = np.zeros(2*local_n_0+1) #this will be the diagonal
    for i in range(2*local_n_0+1):
        D[i] = 4*local_E_C*(-local_n_0+i-local_n_g)**2
    H = np.diag(D) - (local_E_J_max/2)*np.sqrt(np.cos(np.pi*flux/flux_0)**2 + (local_d**2)*np.sin(np.pi*flux/flux_0)**2)*(exp_phi_minus + exp_phi_plus)
    eigenvalues_transmon, eigenvectors_transmon = np.linalg.eigh(H)
    # sort eigenvalues and rearrange corresponding eigenvectors
    # sorted_indices = np.argsort(eigenvalues_transmon)
    # sorted_eigenvalues_transmon = eigenvalues_transmon[sorted_indices]
    # sorted_eigenvectors_transmon = eigenvectors_transmon[:, sorted_indices]
    if local_cutoff_transmon: # slice the wanted subspace
        indices_of_transmon = np.array(range(local_size_of_subspace_T))
        A = np.dot(np.dot(np.linalg.inv(eigenvectors_transmon), H), eigenvectors_transmon)[indices_of_transmon,:][:,indices_of_transmon]
        #print("the H_trans matrix after u_dag*H*U:", A)
        B = np.sum(np.abs(A-np.conj(A.T)))
        print("hermitian of H_trans u_dag*H*U:", B)
        H = np.diag(eigenvalues_transmon[indices_of_transmon])
    else:
        H = np.diag(eigenvalues_transmon)
    return H, eigenvalues_transmon, eigenvectors_transmon

def H_cavity(Wc_local = Wc):
    H = Wc_local*(a_up @ a_down)
    B = np.sum(np.abs(H-np.conj(H.T)))
    print("hermitian of H_cav:", B)
    return H

def H_int_transmon_cavity(g = g, local_n_0 = n_0, local_n_g = n_g, local_cutoff_transmon = False, local_size_of_subspace_T = None, U = None): # also added the chain's dimension
    # U is the eigenvector matrix of the transmon
    D = np.zeros(2*local_n_0+1, dtype=complex) #this will be the diagonal
    for i in range(2*local_n_0+1):
        D[i] = (-local_n_0+i-local_n_g)
    D = np.diag(D)
    if local_cutoff_transmon:
        D = np.dot(np.dot(np.linalg.inv(U), D), U)
        indices_of_transmon = np.array(range(local_size_of_subspace_T))
        D = D[indices_of_transmon,:][:,indices_of_transmon]
        # D = np.where(np.abs(D)<1e-14,0,D)
    trans_cav_kron = np.kron(D,g*(a_down+a_up))
    trans_cav_chain_kron = np.kron(trans_cav_kron, np.identity(2**N, dtype=complex))
    # H = np.kron(g*D,np.kron((a_down+a_up), np.identity(2**N, dtype=complex)))
    B = np.sum(np.abs(trans_cav_chain_kron-np.conj(trans_cav_chain_kron.T)))
    print("hermitian of H_int_trans_cav:", B)
    return trans_cav_chain_kron

def H_Chain(N = N, t = t): # the hamiltonian of the chain alone. N is the number of spins in the system (when we count the up and down spins). t is the tunneling constant
    H_C = np.zeros((2**N,2**N), dtype=complex)
    if N < 2 :
        return np.zeros((2**N, 2**N))
    for i in range(0, N-2, 1):
        term = np.kron(np.identity(2**((i+1)-1), dtype=complex),np.kron(s_plus,np.kron(-s_z, np.kron(s_minus,np.identity(2**(N-((i+1)+2)), dtype=complex)))))
        H_C += -t*(term + term.conj().T) # I am adding +1 in all the dimensions because the range starts from 0 and the spin index start from 1
    for i in range(0,N,1):
        on_site_energy = epsilon_r*np.kron(np.identity(2**((i+1)-1), dtype=complex),np.kron((s_z+0.5*np.identity(2)),np.identity(2**(N-(i+1)),dtype=complex)))
        H_C += on_site_energy
    B = np.sum(np.abs(H_C-np.conj(H_C.T)))
    print("hermitian of H_Chain:", B)
    return H_C

def H_int_transmon_chain(local_n_0 = n_0, N = N, gamma_L = gamma_L, gamma_R = gamma_R, flux = 0, local_cutoff_transmon = False, local_size_of_subspace_T = None, U = None):
    exp_plus = np.exp(1j*np.pi*flux/(2*flux_0)) # should be next to phi_L and -phi_R
    exp_minus = np.conj(exp_plus) # should be next to -phi_L and phi_R
    exp_phi_plus_1 = exp_phi_plus
    if N < 2:
        if local_cutoff_transmon:
            return np.zeros(((local_size_of_subspace_T)*(max_num_photons+1)*2**N, (local_size_of_subspace_T)*(max_num_photons+1)*2**N), dtype=complex)
        else:
            return np.zeros(((2*local_n_0+1)*(max_num_photons+1)*2**N, (2*local_n_0+1)*(max_num_photons+1)*2**N), dtype=complex)
    if local_cutoff_transmon:
        indices_of_transmon = np.array(range(local_size_of_subspace_T))
        exp_phi_plus_1 = np.dot(np.dot(np.linalg.inv(U), exp_phi_plus), U) # basis transformation to the transmon eigenspace
        exp_phi_plus_1 = exp_phi_plus_1[indices_of_transmon,:][:,indices_of_transmon]
    left_term = np.kron(exp_plus*exp_phi_plus_1, np.kron(np.identity((max_num_photons+1), dtype=complex),np.kron(s_minus,np.kron(s_minus, np.identity(2**(N-2), dtype=complex)))))
    right_term = np.kron(exp_minus*exp_phi_plus_1, np.kron(np.identity((max_num_photons+1), dtype=complex), np.kron(np.identity(2**(N-2), dtype=complex),np.kron(s_minus, s_minus))))
    H_int = gamma_L * (left_term + left_term.conj().T) + gamma_R * (right_term + right_term.conj().T)
    B = np.sum(np.abs(H_int-np.conj(H_int.T)))
    print("hermitian of H_int_trans_chain:", B)
    return H_int

def H_total(flux = 0, local_n_g = n_g, local_cutoff_transmon = False, local_size_of_subspace_T = None):
    H_T = H_Transmon(flux=flux,local_n_g=local_n_g,local_cutoff_transmon=local_cutoff_transmon,local_size_of_subspace_T=local_size_of_subspace_T)
    U = H_T[2]
    if local_cutoff_transmon:
        H = np.kron(H_T[0], np.kron(np.identity((max_num_photons+1)), np.identity((2**N))))\
            + np.kron(np.identity(local_size_of_subspace_T), np.kron(H_cavity(), np.identity((2 ** N))))\
            + H_int_transmon_cavity(local_n_g=local_n_g,local_cutoff_transmon=local_cutoff_transmon,local_size_of_subspace_T=local_size_of_subspace_T,U=U)\
            + np.kron(np.identity(local_size_of_subspace_T), np.kron(np.identity((max_num_photons+1)), H_Chain()))\
            + H_int_transmon_chain(flux=flux,local_cutoff_transmon=local_cutoff_transmon,local_size_of_subspace_T=local_size_of_subspace_T,U=U)
    else:
        transmon_part = np.kron(H_T[0], np.kron(np.identity((max_num_photons+1)), np.identity((2**N))))
        cavity_part = np.kron(np.identity((2*n_0+1)), np.kron(H_cavity(), np.identity((2**N))))
        transmon_cavity_int_part = H_int_transmon_cavity(local_n_g=local_n_g,U=U)
        chain_part = np.kron(np.identity((2*n_0+1)), np.kron(np.identity((max_num_photons+1)), H_Chain()))
        transmon_chain_int_part = H_int_transmon_chain(flux=flux,U=U)
        # print("transmon part shape", transmon_part.shape)
        # print("cavity part shape", cavity_part.shape)
        # print("transmon cavity int part shape", transmon_cavity_int_part.shape)
        # print("chain part shape", chain_part.shape)
        # print("transmon chain int part shape", transmon_chain_int_part.shape)
        H = transmon_part\
            + cavity_part\
            + transmon_cavity_int_part\
            + chain_part\
            + transmon_chain_int_part
    B = np.sum(np.abs(H-np.conj(H.T)))
    print("hermitian of H_3:", B)
    return H

```

I will first create the transition amplitude matrix - M.

### Calculate M

```python
amount_of_energies = 30 #eigenvalues_sys.shape[1] #6*2**N
amount_of_energy_diff = np.sum(list(range(amount_of_energies)))

def create_M_and_delta(cutoff=cutoff,size_of_transmon_subspace=size_of_transmon_subspace):
    delta_energy = np.zeros((steps,amount_of_energies,amount_of_energies), dtype=complex) # contains the energy differences, 0 axies in the size of flux array, 1 axies and 2 axies in size of amount of energies. so in each [:,i,j] i will the diff E_i-E_j
    M = np.zeros_like(delta_energy, dtype=complex) # contains the transition probability due to a_down + a_up, here i will put the probabilities. It will contain <i|a_down+a_up|j> in the i'th row and j'th column

    if cutoff:
        operator_2 = np.kron(np.identity(size_of_transmon_subspace), np.kron(g_d*(a_down+a_up), np.identity(2**N))) # that's the operator I want to get the transition probability for
    else:
        operator_2 = np.kron(np.identity((2*n_0+1)), np.kron(g_d*(a_down+a_up), np.identity(2**N))) # that's the operator I want to get the transition probability for

    print("operator should be of total dimension:", operator_2.shape) # sanity test

    for i in range(amount_of_energies): # a loop that iterates from 0 to 6 including
        print("i =", i)
        for j in range(amount_of_energies): # a loop that iterates from 0 to 6 including
            diff = eigenvalues_sys[:,i] - eigenvalues_sys[:,j] # should be an array with number of rows as "steps" (and flux_array) and one column
            delta_energy[:,i,j] = diff
            temp_expectation_vec = np.zeros((steps,1), dtype=complex)
            for step in range(steps):
                vec_i = eigenvectors_sys[step,:,i]
                vec_j = eigenvectors_sys[step,:,j]
                M_ij = vec_i.conj().dot(operator_2.dot(vec_j))# here I want to save the transition probability from j to i for each step because each step have different eigenvectors
                M[step,i,j] = M_ij
    return M, delta_energy

M, delta_energy = create_M_and_delta()
M = np.abs(M)
delta_energy = np.abs(delta_energy)
# delta_energy = np.real(delta_energy)
print(M.shape)
print(delta_energy.shape)
print("are the eigenvalues complex?:",np.iscomplex(eigenvalues_sys).any())
```
